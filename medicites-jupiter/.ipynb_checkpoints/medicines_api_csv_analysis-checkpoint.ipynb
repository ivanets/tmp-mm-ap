```python
import re
import numpy as np

def testStringWithRule(rule, test_data):
    if test_data == None:
        return None
    s = rule.match(test_data)
    if s == None or len(s.groups()) <= 0:
        return None
    if len(s.groups()) == 3:
        (l, d, r) = s.groups()
        result = (l, d, r)
    if len(s.groups()) == 2:
        (l, d) = s.groups()
        result = (l, d, None)
    if len(s.groups()) == 1:
        d = s.groups()
        result = (None, d, None)

for rule in df:
    print(rule, type(rule))
#     rule_re = re.compile(rule.api_re, re.IGNORECASE)
#     (l1, class1, r1) = testStringWithRule(rule_re, rule.test1)
#     (l2, class2, r2) = testStringWithRule(rule_re, rule.test2)
#     (l3, class3, r3) = testStringWithRule(rule_re, rule.test3)
#     print(class1, class2, class3)
    
```
